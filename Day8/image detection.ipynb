```python
import cv2
import matplotlib.pyplot as plt
```

```python
faceClassifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
img = cv2.imread('WhatsApp Image 2022-02-15 at 9.54.35 AM.jpeg')
print(type(img))
faces = faceClassifier.detectMultiScale(img, scaleFactor=1.1, minNeighbors=10, minSize=(100,100))
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),5)
plt.figure(figsize=(15,15))
plt.imshow(img[:,:,::-1])
```

```python
faceClassifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
img = cv2.imread('i.jpeg')
print(type(img))
faces = faceClassifier.detectMultiScale(img, scaleFactor=1.03, minNeighbors=5, minSize=(20,20),maxSize=(45,45))
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
plt.figure(figsize=(15,15))
plt.imshow(img[:,:,::-1])
```
