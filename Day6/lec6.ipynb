#### bright

```python
def bright(img4,beta_value):
    img_brightt=cv2.convertScaleAbs(img4,beta=beta_value)
    return img_brightt
```

```python
img4=cv2.imread("./files/i.jpg")
bright_img=bright(img4,60)
cv2.imshow("bright",bright_img)
```

#### edges

```python
img5=cv2.imread("./files/i.jpg")
print(type(img5))
sobel_y=np.array(
    [
        [-1,-1,-1],
        [0,0,0],
        [1,1,1]

    ],np.float32
)
sobel_x=np.array(
    [
        [-1,0,-1],
        [-1,0,1],
        [-1,0,1]

    ],np.float32
)
sobel=sobel_x +sobel_y
out=cv2.filter2D(img5,-1,sobel,borderType=cv2.BORDER_ISOLATED)
cv2.imshow("sobel",out)
cv2.waitKey(0)
cv2.destroyAllWindows
```

#### sharpening

```python
img6=cv2.imread("./files/i.jpg")
print(type(img6))
def sharpen(img6):
    kernel=np.array([[-1,-1,-1],[-1,-1,-1],[-1,-1,-1]])
    img_sharpen=cv2.filter2D(img6,-1,kernel)
    return img_sharpen
```

```python
out=sharpen(img6)
cv2.imshow("sharben",out)
cv2.waitKey(0)
cv2.destroyAllWindows
```

#### HDR

```python
def HDR(img6):
    hdr=cv2.detailEnhance(img6,sigma_s=12,sigma_r=0)
    return hdr
```

```python
out=HDR(img6)
cv2.imshow("HDR",out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
