# Read from video 

```python
cap = cv2.VideoCapture('Files/11_output.avi')

while (cap.isOpened()):
    stat,frame = cap.read()
    if(stat==True):
        cv2.imshow('frame',frame)
        k=cv2.waitKey(0)
        if k==ord('q'):
            break
        elif k == ord('s'):
            cv2.imwrite('Files/11_output.png', frame)
    else:
        break
cap.release()
cv2.destroyAllWindows()

```

```python
import cv2
import numpy as np

def nothing(x):
    pass

cv2.namedWindow('Image')
cv2.createTrackbar('val','Image',200,255,nothing)
img = np.zeros((500,500,1),np.uint8)

while 1:
    val = cv2.getTrackbarPos('val','Image')
    img[:] = val
    cv2.imshow('Image',img)
    k =cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
import cv2

image = np.zeros((500, 500), np.uint8)

def nothing(x):
    pass

cv2.namedWindow('Image')

cv2.createTrackbar('row', 'Image', 0, image.shape[0]-1, nothing)
cv2.createTrackbar('col', 'Image', 0, image.shape[1]-1, nothing)
cv2.createTrackbar('x', 'Image', 0, image.shape[0]-1, nothing)
cv2.createTrackbar('y', 'Image', 0, image.shape[1]-1, nothing)

while 1:
    row = cv2.getTrackbarPos('row', 'Image')
    col = cv2.getTrackbarPos('col', 'Image')
    x = cv2.getTrackbarPos('x', 'Image')
    y = cv2.getTrackbarPos('y', 'Image')
    
    image = np.zeros((500, 500), np.uint8)
    
    image[row, :] = 255
    image[:, col] = 255

    image[x, y] = 255

    cv2.imshow('Image', image)
    k = cv2.waitKey(1)
    if k == 27:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

def SHOW(img):
    plt.imshow(img)
    print(img.shape)
```

```python
img = np.zeros((500,500,3), np.uint8)

cv2.rectangle(img, (100,100), (320,320), (255, 0, 0), 5)
cv2.line(img, (100,100), (320,320), (255,0,0), 5)
cv2.line(img, (100,320), (320,100), (255,0,0), 5)

cv2.line(img, (210, 320), (210, 100), (255,0,0), 5)
cv2.line(img, (320, 210), (100, 210), (255,0,0), 5)

plt.figure(figsize=(7,7))
plt.imshow(img)

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
